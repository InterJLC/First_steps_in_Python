##### 3.1 Sentencias y bloques

Una sentencia es la unidad de medida más pequeña con sentido propio.

Por otra parte, un bloque es un conjunto de sentencias identadas, es decir, del mismo bloque o grupo de sentencias.

Finalmente un conjunto lo engloban varios bloques.

```python
def maximo(x,y):
    if x>y:
        return x
    else:
        return y

def minimo(x,y):
    if x<y:
        return x
    else:
        return y

#programa principal

x=int(input("Un número: "))
y=int(input("Otro número: "))
print(maximo(x-3, minimo(x+2, y-5)))
```

En este código, nos encontramos con 1 conjunto, que se divide en 3 bloques, y entre todos, suman 14 sentencias.

##### 3.2 Funciones

Una función es una sección del código que es reutilizable, la cual realiza una tarea específica del programa.

Para declarar una función se empieza añadiendo 'def' al código, tal y como se puede apreciar en el código mostrado en el apartado Módulos.

En otro orden, dentro de las funciones se hayarán también variables, las cuales se dividirán en locales y globales. Las variables locales serán las que se encuentren en un bloque o sentencia específica, no afectando al resto de código, y las globales, serán declaradas de forma que puedan reutilizarse en uno o varios bloques o conjuntos.

```python
# Esto es un ejemplo no hace falta ejecutarlo

w = 1
import math
def logaritmo(b,c):
    d = int(b)
    x = int(c)
    return math.log(d,x)
```

En este caso las variables locales serían 'd' y 'x', mientras que la global sería 'w'.

Dentro de las funciones, deberán especificarse siempre un valor con 'return', ya que si no lo hacemos, nos devolverá error en el código, es decir, lo que en una calculadora nos devolvería el famoso 'Syntax Error', aquí en Python el equivalente sería 'None'.

Por otra parte, en lo que respecta a la documentación dentro de las funciones, podremos usar '#' para frases o comentarios de una sola línea, y una triple comilla (""") al inicio y al final de párrafos o textos a ser posible, cortos y concisos explicando información relevante.

Finalmente, en lo que se refiere a la composición de una función, nos encontramos los parámetros, que no son más que las entradas que se le van a dar a la función.

Los hay de 4 tipos: obligatorios, de palabras clave, múltiples o por defecto.

En nuestro ejemplo anterior, los parámetros serían 'b' y 'c' y son de tipo obligatorio.

Importante destacar, que cuando no sabemos el número de parámetros que van a participar en una función, podremos poner delante del mismo '*', es decir, delante del único parametro predefinido un signo de multiplicación para indicar que puede que no sea el único en cuestión.

```python
def PerTri(*a) :
    x=a
    """Documentación: 
    Estamos calculando el perímetro de un triángulo"""
    return (x)
print(PerTri(1+2+3))
```

En este ejemplo en lugar de declarar 3 variables para calcular el perímetro de un triángulo, dejamos que el usuario inserte tantas entradas como desee o necesite.

##### Funciones lampda

##### 3.3 Toma de decisiones

A continuación se van a explicar las palabras reservadas para la toma de decisiones, tales como 'if', 'else' o 'elif'.

Estas condiciones sirven para mostrar si la secuencia o cálculo es cierto (True) o no (False).

```python
def maximo(x,y):
    if x>y:         #Aquí 'if' fija la restricción.
        return x
    else:           #Aquí 'else' da una alternativa.
        return y
print(maximo(10,20))
```

En esta función se muestra el mayor valor entre 'x' e 'y'. 

```python
x = int(input("Porfavor, escriba una cifra: "))
if x < 0:
    x = 0
    print('Negativo')
elif x == 0:
     print('Cero')
elif x == 1:
     print('Unidad')
else:
     print('Valor mayor que uno')
```

En este código 'elif' nos permite tener más de una única alternativa a la restricción presentada por 'if', siendo cerrada por 'else'. También se puede cerrar sin explicaciones con la palabra reservada 'break', que funciona como un "stop".

##### 3.4 Bucles

Los bucles nos permiten ejecutar una serie de sentencias, sean funciones, condiciones u otro tipo de palabras reservadas.

En primer lugar, vamos a ver 'while'. Este bucle funciona ejecutando sentencias hasta que se cumple alguna condición.

```python
contador = 0
while (contador < 9): #El contador 'while' limita el contador hasta el 8.
    print('Cuenta:', contador)
    contador = contador + 1
```

En este código se muestra cómo funciona la función, junto con su restricción.

En segundo lugar tenemos 'for', la cual sirve para iterar entre un abanico de elementos. Ejemplos son:

```python
palabra="unodos"
for w in palabra:
    print(w, len(w))
```

```python
fruits = ['banana', 'apple',  'mango']
for fruit in fruits: 
    print('Current fruit :', fruit)
```

Destacar de nuevo, que en estas palabras también pueden darse casos de anidado.

##### Otras sentencias

Si bien se han explicado las principales sentencias, vamos a repasar algunas de las que también pueden ser frecuentes de ver según qué código.

Estas son: 'break', 'continue', 'pass' y 'range'.

```python
for letter in 'Python': 
    if letter == 'h':
        break
    print('Current Letter :', letter)
```

```python
for letter in 'Python': 
    if letter == 'h':
        continue
    print('Current Letter :', letter)
```

```python
for letter in 'Python':
    if letter == 'h':
        pass
    print('Current Letter :', letter)
```

```python
list(range(5,10))
```

Respectivamente, 'break' provoca una interrupción en el código; 'continue' permite seguir iterando respetando la restricción; 'pass' directamente no hace nada, ni establece ninguna restrcción; y por último, 'range' se puede utilizar para mostrar una secuencia de números.

##### Programación funcional

Se dan otros métodos en los que no sólo es recibir y ejecutar una orden directa, sino que trabajan como 'subfunciones' que se ejecutan en base a una lista. Los más destacados son 'filter', 'map' y 'reduce'.

```python
def es_par (x):
    if x % 2 == 0:
        return True
    
numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9]

print(list(filter(es_par, numeros)))
```

```python
def cube (x): 
    if x*x*x:
        return x*x*x

a = list(map(cube, range(1, 10)))

print(a)
```

```python
from functools import reduce
```

```python
def suma(a,b):
    return a + b

print(reduce(suma,[1, 2, 3, 4])) 
```

Además, la función print puede controlar y manipular el formato de salida de la cadena o resultado a mostrar:

```python
print('Art: %5d, Precio por Unidad: %8.2f' % (453, 59.058))
print('Art: %4d, Precio por Unidad: %7.2f' % (453, 59.058))
print('Art: %3d, Precio por Unidad: %6.2f' % (453, 59.058))
print('Art: %4d, Precio por Unidad: %6.1f' % (453, 59.058))
print('Art: %3e, Precio por Unidad: %6.1f' % (453, 59.058))
print('Art: %3E, Precio por Unidad: %6.1f' % (453, 59.058))
print('Art: %3g, Precio por Unidad: %6.1f' % (453, 59.058))
print('Art: %s, Precio por Unidad: %6.1f' % (453, 59.058))
print("%+d"% (42))
print("% d"% (42))
print("%+2d"% (42))
print("% 2d"% (42))
print("%2d"% (42))
```

En este sentido, la nomenclatura utilizada es: d = entero; f = flotante; X1.X2 refleja numero de decimales y espacios por la izquierda respectivamente.

Otros formatos son:
- d i ; número entero con signo.
- u ; número entero sin signo.
- e ; flotante en formato exponencial (minúscula).
- E ; ídem pero mayúscula.
- F f ; coma flotante en formato decimal.
- g G ; como 'e' si exponente > -4 o < precisión; como 'f' en otro caso. (Ver PW T5 p.11).*
- c ; carácter único.
- r ; cadena (usando repr).
- s ; cadena (usando str).

#### 5.2 Excepciones

Las excepciones son errores que surgen cuando el código no contempla ciertas situaciones que se pueden generar.

Estos errores los podemos 'atrapar' para que cuando se den, nuestro código comtemple una salida acorde a es error o excepción específico.

```python
dividendo = float(input('Introduce dividendo: '))
divisor = float(input('Introduce divisor: '))
Resultado = dividendo / divisor
print('El resultado es ', Resultado)
```

Como se puede observar, en este código nos sale un error con el nombre 'ZeroDivisionError'. Este error los podemos capturar y hacer que el código muestre al usuario una respuesta más 'amigable' de la siguiente manera:

```python
dividendo = float(input('Introduce dividendo: '))
divisor = float(input('Introduce divisor: '))
try:
    Resultado = dividendo / divisor
    print('El resultado es ', resultado)
except ZeroDivisionError: #sólo estamos capturando ese tipo de error.
    print('No se permite la división por 0')
```

Aqí podemos ver que el programa en lugar de dar un error de la anterior forma, muestra una respuesta personalizada del error.

```python
try:
    dividendo = float(input('Introduce dividendo: '))
    divisor = float(input('Introduce divisor: '))
    
    Resultado = dividendo / divisor
    print('El resultado es ', resultado)
except:
    print('División no válida')
```

De esta otra forma, se pueden capturar todos los errores de este apartado, ya que sólo contempla 2 opciones: o se ejecuta bien (try) o no se ha ejecurado bien (except).

```python
try:
    dividendo = float(input('Introduce dividendo: '))
    divisor = float(input('Introduce divisor: '))
    Resultado = dividendo / divisor
    print('El resultado es ', resultado)
except ValueError: #solo estamos capturando este tipo de error
    print('Solo valen números')
except:
    print('División no válida')
```

En este último caso, vemos como podemos utilizar varios 'try' y 'except' de igual forma que utilizamos los comandos 'if' y 'else', por ejemplo.

### Tema 10: Web Scrapping

Denominamos 'web scrapping' a la extracción de información de páginas web de internet.

En estas webs podremos extraer información de su propia estructura, de ficheros (.CSV, .XLSX, etc) o a través de APIs (interfaces de aplicación).

En este tema nos vamos a centrar en extraer información de webs y tratarla con la librería Pandas. Para ello vamos a utilizar adicionalmente la librería 'BeautifulSoup'.

Esta librería nos permitirá extraer información.

Cuando visitamos una web se realiza una solicitud denominada 'GET'. Después de que esto ocurra, el servidor nos devuelve la solicitud con el contenido de un fichero que dice como mostrar la web.

Estos ficheros suelen ser .HTML, .CSS, .JS, etc. Incluso se añaden imágenes en su respectivo formato.

En temas de web scrapping sólo debemos fijarnos en el código .HTML, ya que es donde se organiza y se localiza el contenido escrito de la web.

En los códigos .HTML se utilizan etiquetas, las cuales nos permiten saber dónde empieza una un elemento u otro.

Es decir, generalmente las etiquetas deberan 'abrisrse' y 'cerrarse': 

```python
# No ejecutar

<html> ... </html>
```

Sin embargo, con la nueva generación de HTML5, no es necesario cerrar las etiquetas, ya que al comenzar una nueva se da por hecho que se cierra la anterior. 

```python
# Aquí vemos como se estructuraría un elemento con título y cuerpo:

<html>
   <head>
   </head>
   <body>
   </body>
</html>
```

```python
# Aquí veríamos como se desarrollaría unas lineas de texto:

<html>
       <head>
       </head>
       <body>

       <p>Primer párrafo de texto...
        <a href="https://www.parrafodetxt.es">Mi primer párrafo web</a>
        </p>

       <p>Segundo párrafo de texto...</p>
       
       </body>
</html>

# Con la etiqueta 'a' incluímos webs dentro de nuestro código HTML.
```

Dentro de las etiquetas HTML podemos encontrar en orden de relevancia o categoría las etiquetas: HTML sería la etiqueta padre de HEAD por ejemplo. BODY lo sería P también.

Después nos referiríamos a los sibling, las cuales están anidadas en una misma etiqueta. Por ejemplo HEAD y BODY están dentro de la etiqueta HTML con el mismo rango, por lo que son siblings.

Por último están las etiquetas CHILD, las cuales marcan los elementos que están dentro de otros. Ene ste caso P está dentro de BODY, luego P es de tipo CHILD.

Bien, pues lo primero que se debe hacer antes de analizar una web es descargarla. Para ello vamos a utilizar la librería 'Requests' de Python.

Esta biblioteca realizará la solicitud 'GET' al servidor web y descargará el contenido 'HTML'.

Luego de descargarla, procederemos a analizarla con la librería nates mencionada: BeautifulSoup.
